```python
%matplotlib inline

# imports
import os
import ast
import pickle
import numpy as np
import pandas as pd
import tensorflow as tf
import keras.optimizers
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, Conv2DTranspose, UpSampling2D, Cropping2D
from keras.models import Model
from keras.datasets import mnist
from keras import backend as K

# plotting
import matplotlib
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=15)
plt.rcParams['figure.figsize'] = [20, 15]

# prevent the tensorflow warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
from tensorflow.python.util import deprecation
deprecation._PRINT_DEPRECATION_WARNINGS = False
```

# 1. Classification

```python
# load in the dataset
dataset = tf.keras.datasets.mnist

# split the dataset into training and testing data
(x_train, y_train), (x_test, y_test) = dataset.load_data()

# normalise the data
x_train = x_train / 255.0
x_test = x_test / 255.0

print(x_test.shape)
```

```python
# plot some images
vmax = np.max(np.abs(x_train))
vmin = -np.max(np.abs(x_train))
f, sub = plt.subplots(5,5)
for i in range(5):
    for j in range(5):
        k = i + j
        sub[i,j].imshow(x_train[k,:,:], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
        sub[i,j].set_xticklabels([])
        sub[i,j].set_yticklabels([])
plt.savefig('images.pdf')
```

```python
# build the neural network
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='sigmoid'),
    tf.keras.layers.Dense(10, activation='sigmoid')
])
```

```python
# complile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# train the model
model.fit(x_train, y_train, epochs=2)
```

```python
# evaluate the model
model.evaluate(x_test, y_test, verbose=2)
```

```python
# predict for given input
N = 4 # number to predict
x_pred = x_test[:N,:,:]
predictions = model.predict(x_pred)
```

```python
# plot the predictions
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
ax1.imshow(x_pred[0,:,:], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
ax2.imshow(x_pred[1,:,:], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
ax3.imshow(x_pred[2,:,:], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
ax4.imshow(x_pred[3,:,:], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
ax1.title.set_text('prediction: {}'.format(np.argmax(predictions[0])))
ax2.title.set_text('prediction: {}'.format(np.argmax(predictions[1])))
ax3.title.set_text('prediction: {}'.format(np.argmax(predictions[2])))
ax4.title.set_text('prediction: {}'.format(np.argmax(predictions[3])))
ax1.set_yticklabels([])
ax2.set_yticklabels([])
ax3.set_yticklabels([])
ax4.set_yticklabels([])
ax1.set_xticklabels([])
ax2.set_xticklabels([])
ax3.set_xticklabels([])
ax4.set_xticklabels([])
plt.savefig('predictions.pdf')
```

# 2. Autoencoders

```python
(x, _), (_, _) = mnist.load_data()
x = x.astype('float32') / 255.
x = np.reshape(x, (len(x), 28, 28, 1))
image_dimention = x.shape[-2]
print(x.shape)
```

```python
# hyperparameters
count = 40000
validation_split = 0.20
optimizer = 'adadelta'
learning_rate = 1.0
epochs = 10
batch_size = 128
loss = 'binary_crossentropy'
shuffle = True
```

```python
# define the autoencoder
input_img = Input(shape=(image_dimention, image_dimention, 1))
encoded = Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
encoded = MaxPooling2D((2, 2), padding='same')(encoded)
encoded = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
encoded = MaxPooling2D((2, 2), padding='same')(encoded)
encoded = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
encoded = MaxPooling2D((2, 2), padding='same')(encoded)
latent_shape = (int(encoded.shape[1]), int(encoded.shape[2]), int(encoded.shape[3]))
print(latent_shape)
decoded = Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
decoded = UpSampling2D((2, 2))(decoded)
decoded = Conv2D(8, (3, 3), activation='relu', padding='same')(decoded)
decoded = UpSampling2D((2, 2))(decoded)
decoded = Conv2D(16, (3, 3), activation='relu')(decoded)
decoded = UpSampling2D((2, 2))(decoded)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(decoded)
autoencoder = Model(input_img, decoded)
```

```python
# print the model architecture
autoencoder.summary()
```

```python
# define the encoder
encoder = Model(input_img, encoded)
```

```python
# define the decoder
encoded_input = Input(shape=latent_shape)
decoder_layers = len(autoencoder.layers) - len(encoder.layers)
for i in range(decoder_layers, 0, -1):
    decoder_layer = autoencoder.layers[-i]
    if i == decoder_layers:
        nest = decoder_layer(encoded_input)
    else:
        nest = decoder_layer(nest)
decoder = Model(encoded_input, nest)
```

```python
# compile the autoencoder
if optimizer == 'adam':
    opt = keras.optimizers.Adam(lr=learning_rate)
if optimizer == 'sgd':
    opt = keras.optimizers.SGD(learning_rate=learning_rate, momentum=0.1)
if optimizer == 'adadelta':
    opt = keras.optimizers.Adadelta(learning_rate=learning_rate)
autoencoder.compile(optimizer=opt, loss=loss)
latent_shape = tuple(encoder.layers[-1].output_shape[1:4])
```

```python
# train the autoencoder
training = autoencoder.fit(x, x, validation_split=validation_split, epochs=epochs, batch_size=batch_size, shuffle=shuffle)
```

```python
# plotting the training
train_loss = np.array(training.history['loss'])
test_loss = np.array(training.history['val_loss'])
plt.plot(train_loss, label='train loss')
plt.plot(test_loss, label='test loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend()
plt.savefig('training.pdf')
```

```python
# encode and then decode some test data
encoded_imgs = encoder.predict(x)
decoded_imgs = decoder.predict(encoded_imgs)
```

```python
# plot the decoded test data
n = 10
vmax = np.max(np.abs(x))
vmin = -np.max(np.abs(x))
plt.figure(figsize=(10, 2), dpi=100)
for i in range(10):
    ax = plt.subplot(2, n, i+1)
    plt.imshow(x[i,:,:,0], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
    plt.tick_params(axis='both', which='both', bottom='off', top='off', labelbottom='off', right='off', left='off', labelleft='off')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax = plt.subplot(2, n, i+n+1)
    plt.imshow(decoded_imgs[i,:,:,0], norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
    plt.tick_params(axis='both', which='both', bottom='off', top='off', labelbottom='off', right='off', left='off', labelleft='off')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
plt.savefig('predictions.pdf')
```

```python
# plot the latent spaces
n = 10
vmax = np.max(np.abs(encoded_imgs))
vmin = -np.max(np.abs(encoded_imgs))
plt.figure(figsize=(10, 10), dpi=100)
for i in range(10):
    ax = plt.subplot(n, 1, i + 1)
    plt.imshow(np.reshape(encoded_imgs[i,:,:,:],(latent_shape[0],latent_shape[1]*latent_shape[2])), norm=plt.Normalize(0,vmax), cmap=matplotlib.cm.binary, interpolation='hamming')
    plt.tick_params(axis='both', which='both', bottom='off', top='off', labelbottom='off', right='off', left='off', labelleft='off')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
plt.savefig('latent.pdf')
```

```python
# plot the flat latent spaces
n = 10
shift = np.max(encoded_imgs)
plt.figure(figsize=(10, 10), dpi=100)
for i in range(10):
    plt.plot(np.reshape(encoded_imgs[i,:,:,:],(latent_shape[0]*latent_shape[1]*latent_shape[2]))+i*shift, label='${0}$'.format(i), linewidth=2.0)
plt.savefig('latent_flat.pdf')
```
